#### Generate 1000 random points
#### Build a Linear Regression model from scratch
#### Use Gradient descent to find best fit line(slope,intercept) with least square criterion

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=1000
np.random.seed(42)
x=np.random.randn(n)
d=10
y=(d*np.random.randn(n))+x
```

```python
class GradientDescentRegression:
    def __init__(self,learning_rate=0.01,epochs=500):
        self.learning_rate,self.epochs=learning_rate,epochs
    def fit(self,x,y):
        m=5
        c=0
        n=x.shape[0]
        for i in range(self.epochs):
            y_pred=m*x+c
            m_gradient=(-2/n)*(np.sum(x*(y-y_pred)))
            c_gradient=(-2/n)*(np.sum(y-y_pred))
            m=m-(self.learning_rate*m_gradient)
            c=c-(self.learning_rate*c_gradient)
        self.m,self.c=m,c
        return m,c
    def predict(self,x):
        y=self.m*x+self.c
        return y
```

```python
lr=GradientDescentRegression()
m,c=lr.fit(x,y)
y_pred=lr.predict(x)
```

```python
plt.scatter(x,y,c='black')
plt.plot(x,y_pred)
```
